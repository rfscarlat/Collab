```python
import matplotlib.pyplot as plt

def printCollatz(n):
  x_coord = []
  y_coord = []
  i = 0
  while n != 1:
    i = i + 1
    x_coord.append(i)
    y_coord.append(n)
  #  print(n, end = ' ')
    if n & 1:
      n = 3 * n + 1
    else:
      n = n // 2
  #print(n) # print 1 at the end
  x_coord.append(i+1)
  y_coord.append(1)
  plt.plot(x_coord, y_coord)
  plt.show()
  print("\nSteps = ",max(x_coord))
  print(' '.join(str(x) for x in y_coord))


# Driver code
printCollatz(27)
```
